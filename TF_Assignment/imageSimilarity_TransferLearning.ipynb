```python
import tensorflow as tf
import numpy as np
import time
from PIL import Image
from datetime import timedelta
import os
import inception
tf.__version__
import prettytensor as pt
```

```python
inception.maybe_download()
```

```python
model = inception.Inception()
from inception import transfer_values_cache
```

```python
lst=[]
for i in range(1,11):
    img1=Image.open("images/"+str(i)+".jpg").resize((224,224))
    img1_np=np.array(img1)
    lst.append(img1_np)
```

```python

imgs=np.array(lst)
imgs.shape
```

```python
file_path_cache_train = 'inception_img1.pkl'
```

```python
transfer_values_train = transfer_values_cache(cache_path=file_path_cache_train,images=imgs,model=model)
```

```python
transfer_values_train.shape
```

```python
from sklearn.metrics.pairwise import pairwise_distances
```

```python
from IPython.display import Image as im
import matplotlib.pyplot as plt
import pylab
for i in range(0,9):
    plt.figure()
    plt.imshow(lst[i])
    plt.figure()
    plt.imshow(lst[i+1])
    pylab.show()
    val=pairwise_distances(transfer_values_train[i].reshape(1,-1),transfer_values_train[i+1].reshape(1,-1),metric='cosine')
    print(val)
    

```
